On cherche à résoudre une équation différentielle de la forme $y''+b\times y' +c\times sin(y)=0$

```python
%matplotlib notebook

from scipy.integrate import odeint
import math
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def der (y,t,b,c):
    y,yp=y
    ypp=-b*yp-c*math.sin(y)
    return [yp,ypp]
```

```python
b=0.25
c=5.0
t=np.linspace(0,10,100)
```

```python
Y0=[math.pi/2,0]
```

```python
sol=odeint(der,Y0,t,args=(b,c))
```

```python
plt.plot(t, sol[:, 0], 'b', label='y\'') #on trace la courbe en bleu 'b'
```

```python
plt.plot(t, sol[:, 1], 'g', label='y\'\'')
```

```python
plt.legend(loc='best')
```

```python
plt.xlabel('t')
```

```python
plt.grid()
```
